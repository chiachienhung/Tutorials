### **2. Density**
Describe the portion of the **potential connections** (i.e. the person could know the others) in a network that are **actual connections** (i.e. the person knows). The density is 0 for a graph without edges and 1 for a complete graph. The denominator of equation (2a) and (2b) is the maximum number of edges in the graph $G$.


  - **Undirected**: since edge $e_{ij}$ is the same as $e_{ji}$, here $n$ is the number of nodes and $m$ is the number of edges in undirected graph $G$.
  $$d_{undirected} = \frac{Actual\_Connections}{Potential\_Connections} = \frac{m}{(\frac{n(n-1)}{2})} \tag{2a}$$

  - **Directed**: here $n$ is the number of nodes and $m$ is the number of edges in directed graph $G$.
  $$d_{directed} = \frac{Actual\_Connections}{Potential\_Connections} = \frac{m}{n(n-1)} \tag{2b}$$

````
## Density for undirected graph
print("Density of the undirected graph with weights:", nx.density(Actor_nw_sym_weight))
print("Density of the undirected graph without weights:", nx.density(Actor_nw_sym))
## Density for directed graph
print("Density of the directed graph:", nx.density(Actor_nw_asym))
````

### **3. Clustering Coefficient**

A measure of the degree to which nodes in a graph tend to cluster together. Here we will only discuss the local clustering coefficient, since we care which node in a network has a tendency to from clusters. **Local Clustering Coefficient** of a node in a network is the fraction of pairs of the node’s neighbours that are adjacent to each other.

 - **Undirected-unweighted**: the clustering of a node $u$ is the fraction of possible triangles through that node that exist, where $T(u)$ is the number of edges among the neighbors of node $u$. And $neighbor(u)$ is the number of neighbor nodes for node $u$.
    - If $neighbor(u)>1$ 
  $$c_{u} = \frac{\text{Actual Connections (neighbor nodes)}}{\text{Potential Connections (neighbor nodes)}} = \frac{T(u)}{(\frac{neighbor(u)(neighbor(u)-1)}{2})} \tag{3a}$$
    - else
    $$c_u = 0 \tag{3b}$$ 
  
- **Undirected-weighted**: there are several ways to define clustering. The one here is defined as the geometric average of the subgraph edge weights. Here $neighbor(u)$ is the number of neighbor nodes for node $u$ and the edge weights $\hat{w}_{uv}$ are normalized by the maximum weight in the network $\hat{w}_{uv}=w_{uv}/max(w)$.
    - If $neighbor(u)>1$ 
  $$c_u = \frac{\text{Actual Connections (neighbor nodes)}}{\text{Potential Connections (neighbor nodes)}} = \frac{\sum_{vw} (\hat{w}_{uv}\hat{w}_{uw} \hat{w}_{vw})^{1/3}}{(\frac{neighbor(u)(neighbor(u)-1)}{2})} \tag{3c}$$

    - else
    $$c_u = 0 \tag{3d}$$ 

- **Directed**: See more [here](https://networkx.github.io/documentation/stable/reference/algorithms/generated/networkx.algorithms.cluster.clustering.html#networkx.algorithms.cluster.clustering)

### **4. Average Clustering**

Average clustering coefficient for the network, where $n$ is the number of nodes in the network $G$ and $c_u$ is the clustering coefficient of the node $u$ in $G$:

$$C = \frac{1}{n}\sum_{u \in G} c_u \tag{4}$$

#### **More advanced...**

Below is more advanced explanation of how to calculate the clustering coefficient of undirected graph in both weighted and unweighted scenario.

````
# For undirected weighted graph
# Modified from here: https://networkx.github.io/documentation/stable/_modules/networkx/algorithms/cluster.html#clustering
def _weighted_triangles_and_degree_iter(G, nodes=None, weight='weight'):
    """ Return an iterator of (node, degree, weighted_triangles).

    Used for weighted clustering.

    """
    if weight is None or G.number_of_edges() == 0:
        max_weight = 1
    else:
        max_weight = max(d.get(weight, 1) for u, v, d in G.edges(data=True))
    if nodes is None:
        nodes_nbrs = G.adj.items()
    else:
        nodes_nbrs = ((n, G[n]) for n in G.nbunch_iter(nodes))

    def wt(u, v):
        #print(G[u][v].get(weight, 1), max_weight)
        return G[u][v].get(weight, 1) / max_weight

    for i, nbrs in nodes_nbrs:
        inbrs = set(nbrs) - {i}
        weighted_triangles = 0
        seen = set()
        for j in inbrs:
            seen.add(j)
            # This prevents double counting.
            jnbrs = set(G[j]) - seen
            # Only compute the edge weight once, before the inner inner
            # loop.
            print('Max Weight:', max_weight)
            for k in inbrs & jnbrs:
              print('Triangular Nodes:', i, j, k)
              print('Weight:', G[i][j].get(weight, 1), G[j][k].get(weight, 1), G[k][i].get(weight, 1))
              print('Normalized weight:', wt(i, j), wt(j, k), wt(k, i))
            wij = wt(i, j)
            weighted_triangles += sum((wij * wt(j, k) * wt(k, i)) ** (1 / 3)
                                      for k in inbrs & jnbrs)
        yield (i, len(inbrs), weighted_triangles)
````

````
# For undirected unweighted graph
# Modified from here: https://networkx.github.io/documentation/stable/_modules/networkx/algorithms/cluster.html#clustering
from collections import Counter
def _triangles_and_degree_iter(G, nodes=None):
    if nodes is None:
        nodes_nbrs = G.adj.items()
    else:
        nodes_nbrs = ((n, G[n]) for n in G.nbunch_iter(nodes))

    for v, v_nbrs in nodes_nbrs:
        vs = set(v_nbrs) - {v}
        gen_degree = Counter(len(vs & (set(G[w]) - {w})) for w in vs)
        ntriangles = sum(k * val for k, val in gen_degree.items())/2
        yield (v, len(vs), ntriangles, gen_degree)
````

````
# Calculate clustering coefficient for undirected graph in both weighted & unweighted case
# Modified from here: https://networkx.github.io/documentation/stable/_modules/networkx/algorithms/cluster.html#clustering
def clustering_undirected(G, nodes=None, weight=None):
    assert G.is_directed()==False, 'The graph must be an undirected graph!'
    if weight is not None:
        td_iter = _weighted_triangles_and_degree_iter(G, nodes, weight)
        clusterc = {v: 0 if t == 0 else 2*t / (d * (d - 1)) for
                    v, d, t in td_iter}
    else:
        td_iter = _triangles_and_degree_iter(G, nodes)
        clusterc = {v: 0 if t == 0 else 2*t / (d * (d - 1)) for
                    v, d, t, _ in td_iter}
    if nodes in G:
        # Return the value of the sole entry in the dictionary.
        return clusterc[nodes]
    return clusterc
````

### **5. Average Shortest Path Length**

The average shortest path length is the sum of the shortest path (-or distance, geodesic path) $d(u,v)$ between a set of nodes $S$ (assuming the length is zero if $v$ is not reachable from $v$) normalized by $n(n-1)$ where $n$ is the number of nodes in network $G$. 

If ```weight``` is None, every edge has weight/distance/cost 1. If a string, use this edge attribute as the edge weight. Any edge attribute not present defaults to 1.

$$<d> =\sum_{u,v \in S} \frac{d(u, v)}{n(n-1)} \tag{5}$$ 

````
print('Average shortest connection between actors (unweighted):', nx.average_shortest_path_length(Actor_nw_sym_weight, weight=None))
print('Average shortest connection between actors (weighted):', nx.average_shortest_path_length(Actor_nw_sym_weight, weight="weight"))
````

### **Prominance**

Find out who are the most important nodes in a social network?

* **Centrality**: check which node (i.e. friend) is involved in many edges (i.e. relationship)
  - **Degree Centrality**: the fraction of nodes it is connected to. Degree of node $u$ divided by the maximal possible degree (i.e. number of nodes in the graph subtract by 1)

  $$C_D(u) =\frac{deg(u)}{n-1}\tag{undirected}$$
  $$C_D(u) =\frac{deg^{out}(u)}{n-1}\tag{directed}$$

  - **Betweenness Centrality**: how many pairs of individuals would have to go through you in order to reach one another in the minimum number of hops. Betweenness centrality of a node $v$ is then defined as the sum of the fraction of all-pairs shortest paths that pass through $u$. Here $V$ is the set of nodes, $\sigma(s, t)$ is the number of shortest $(s,t)$-paths, and $\sigma(s, t|u)$ is the number of those paths passing through some node $u$ other than $s,t$.
$$C_B(u) =\sum_{s,t \in V} \frac{\sigma(s, t|u)}{\sigma(s, t)}$$

      - **Normalized**: $n$ is the number of nodes in the graph

  $$C^{'}_B(u) = \frac{C_B(u)}{(\frac{(n-1)(n-2)}{2})} \tag{undirected}$$
  $$C^{'}_B(u) = \frac{C_B(u)}{(n-1)(n-2)} \tag{directed}$$

  - **Closeness Centrality**: focus on how close a node $u$ is to all the other nodes in the network. It is based on the length of the average shortest path between a node and all the other nodes (i.e. $k-1$ reachable nodes) in the graph. Here $d(u, v)$ is the shortest-path distance between $u$ and $v$. 
  $$C_C(u) = \frac{1}{\sum_{v=1}^{k-1} d(u, v)}$$

      - **Normalized**: $k-1$ is the number of nodes that can reach $u$. 

  $$C^{'}_C(u) = \frac{k - 1}{\sum_{v=1}^{k-1} d(u, v)}$$




* **Prestige**: Take the direction of arc into account (i.e. directed graph)
  - **Degree Prestige / Popularity**: the node which is more prestigious when having more other nodes point to it (i.e. higher in-degree value). It is considered as the fraction of nodes its incoming edges are connected to. In-degree of node $u$ divided by the maximal possible in-degree (i.e. number of nodes in the graph subtract by 1) ```nx.in_degree_centrality(DiGraph)```

  $$C_{D_{in}}(u) =\frac{deg^{in}(u)}{n-1}$$

  - **Proximity Prestige**: Measure based on the distance of input domains. Input domain of a node in a directed graph is the number or percentage of all other nodes that are connected by a path to this node. Direct nominations (i.e. choices) should count more than indirect nominations. Here $n$ is the number of nodes in the graph and $d(i, u)$ is the distance between node $i$ and node $u$.
   $$C_P(u) =\frac{\text{fraction of all nodes that are in u's input domain} }{\text{average distance from u to other nodes in the input domain}} = \frac{\frac{|I_u|}{(n-1)}}{\frac{\sum_{i \in I_u} d(i, u)}{|I_u|}}$$

  - **Page Rank**: more prestigious if a node has lots of other prestigious people in the input domain. [Check more here](https://networkx.github.io/documentation/stable/reference/algorithms/link_analysis.html)
